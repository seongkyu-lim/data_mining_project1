제가 사용하고자하는 apriori 함수를 사용하려면 transactions를 list형으로 사용해야하여 아래코드를 통해 변형시켜주었습니다.

```
transactions = []
for i in range(0, len(df_winningnum)):
    transactions.append([str(df_winningnum.values[i,j]) for j in range(0, 7)])

print(transactions)
```

#problem 1

python의 apyori 라이브러리에 있는 apriori 함수를 사용하여 One-antecedent-one-consequent rules를 구하였습니다.

```
!pip install apyori
```

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

앞서 구축한 transactions와 함께 min_support, min_confidence, min_lift, max_length(itemset안에 item최대 개수)를 선언하여 그해 부합하는 rules들을 구하였습니다.
<br/>min_support = 0.005 (대략 939회의 추첨 중 5번의 추첨에서 숫자가 나왔을 경우)
<br/>min_confidence = 0.2 (20%)
<br/>min_lift = 2.7
<br/>max_length = 2 (One-antecedent-one-consequent rules)
<br/>min_length = 2

```
association_rules1 = apriori(transactions, min_support=0.005, min_confidence=0.2, min_lift=2.7, max_length=2) #  setting min_support, max item lengthin itemset
association_results1 = list(association_rules1)

association_results1 = list(filter(lambda x: len(x.items) > 1, association_results1)) # item num in itemset is bigger than 1
```

```
print(len(association_results1))
```

```
print(association_results1[0])
```




Apriori library를 사용하게 되면

결과 값이 frozenset이라는 자료형으로 저장이 됩니다.

위의 결과값 첫번째 rule을 출력해보면 frozenset의 구성을 확인할 수 있고 인덱싱을 통해

각 rule과 support, confidence lift값을 아래와 같이 출력하였습니다.

```
for item in association_results1:

    print("Rule: " + str(item[2][0][0]) + " -> " + str(item[2][0][1]))
    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("----------------------------------------------")
```

# problem 3

2번 문제와 같이 가장 함께 선택된 빈도가 많은 숫자 집합일수록 support값이 높습니다.

따라서 apriori 알고리즘을 통해 얻어진 3개의 숫자를 가진 itemset들 중 support 값이 가장 큰 값을 찾아보았습니다.

```
association_rules2 = apriori(transactions, min_support=0.008, max_length=3) #  setting min_support, max item lengthin itemset
association_results2 = list(association_rules2)

association_results2 = list(filter(lambda x: len(x.items) > 2, association_results2)) # item num in itemset is bigger than 2
```

```
print(len(association_results2))
```

```
print(association_results2)
```

```
for item in association_results2:

    print("Itemsets: " + str(item[0]))
    print("Support: " + str(item[1]))
```

min_support값을 여러값으로 조정하여 알고리즘을 실행해본 결과, 0.008을 넘는 세개의 item을 가진 itemset은 하나 뿐이며

## **Itemset = {6, 18, 31}** 
이 그에 해당합니다.
